### <span style='color:darkviolet'> Data Discription</span>

- There are 641 rows and 19 columns in the dataset.
- The columns represent different months, seasons, and aggregate rainfall values for each district.
- Monthly rainfall is recorded in columns from JAN to DEC.
- ANNUAL column sums up the total annual rainfall.
- Seasonal aggregates are provided for Jan-Feb, Mar-May, Jun-Sep, and Oct-Dec.
- Rainfall amounts vary significantly, with annual totals ranging from approximately 2805 mm to over 4000 mm.

### <span style='color:darkviolet'>Columns Insights

- **STATE_UT_NAME** : Name of the state or union territory.
- **DISTRICT**: Name of the district within the state/UT.
- **JAN to DEC**: Average monthly rainfall (in mm) from January to December.
- **ANNUAL**: Total annual rainfall (in mm) for each district.
- **Jan-Feb**: Total rainfall (in mm) for January and February.
- **Mar-May**: Total rainfall (in mm) for March, April, and May.
- **Jun-Sep**: Total rainfall (in mm) for June, July, August, and September.
- **Oct-Dec**: Total rainfall (in mm) for October, November, and December.

### <span style='color:darkviolet'>Outliers Tretment</span>

```python
df
```

```python
for i in df.select_dtypes(include = ['int','float']):
    sns.boxplot(data=df,x=i)
    plt.show()
```

```python
def outlier_treatment(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3-Q1
    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR

    ub = df[col]>upper_limit
    lb = df[col]<lower_limit

    df.loc[ub,col] = df[col].median()
    df.loc[lb,col] = df[col].median()

    return df
```

```python
for j in df.select_dtypes(['int','float']):
    outlier_treatment(j)
```

```python
for i in df.select_dtypes(include=['int','float']):
    sns.boxplot(data=df,x=i)
    plt.show()
```

### <span style='color:brown'>Observation

- In our dataset some outliers present in some columns.
- We used IQR Method to reduce the outliers.
